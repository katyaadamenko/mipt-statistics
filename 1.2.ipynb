```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
n = 10000
theta = 1
```

```python
from math import factorial
```

```python
def find_estimate(sample, k):
    s = np.arange(1, sample.size + 1)
    return (factorial(k) * np.ones(sample.size) / ((sample ** k).cumsum() / s) ) **  (1. / k)
```

```python
sample = sps.expon.rvs(loc = 0, scale = 1 / theta, size = n)
```

Построим график зависимости $\left|\sqrt[k]{\frac{k!}{\overline{X^k}}} - \theta\right|$ от k:

```python
colors = np.array(['indigo', 'coral', 'red', 'chocolate', 'grey', 'blue', 'deeppink', 'lime', 'orange', 'deepskyblue', 'g', 
          'darkorchid', 'yellow', 'sandybrown', 'black'])
s = np.arange(1, n + 1)
plt.figure(figsize=(20, 10))
for i in range(1, colors.size + 1):
    plt.plot(s, abs(find_estimate(sample, i) - theta), color = colors[i - 1], linewidth=2, label= 'k = ' + str(i))
plt.legend()
plt.xlim((8000, n))
plt.ylim((0, 0.3))
plt.xlabel('k', fontsize = 20)
plt.ylabel('')
plt.title(r'$\theta$ = 1')
plt.grid() 

plt.show()
```

Заметим, что при n близких к N = 10000 наименьший модуль разность оценки и истинного значения $\theta$ достигается при k = 1, k = 2 и k = 3. Затем при увеличении k разность увеличивается.
