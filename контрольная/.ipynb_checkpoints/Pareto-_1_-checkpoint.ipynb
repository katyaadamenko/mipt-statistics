```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import pareto
%matplotlib inline
```

### Сгенерируем выборку объёма 1000 из распределения Парето

```python
c = 3

mean, var, skew, kurt = pareto.stats(c, moments='mvsk') ## Сохраняем моменты

pareto_rv = sts.pareto(c)
sample = pareto_rv.rvs(size=1000)

print mean, var
```

### Построим гистограмму выборки и нарисуем поверх нее теор. плотность распределения 

```python
x = np.linspace(0,4,100)
pdf = pareto_rv.pdf(x)
plt.plot(x, pdf, lw=3)

plt.hist(sample, bins=50, normed=True)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
sample_number = 1000

sample_5 = [] ##Создаем массив, в который будем записывать выборочные средние

## Цикл последовательно генерирует 1000 выборок заданного объема
for i in range(sample_number):
    tmp = pareto_rv.rvs(size=5) ## Генерируем выборку объема 5
    sample_5.append(np.mean(tmp)) ## Записываем выборочное среднее выборки в массив

mu_5 = mean
sigma_5 = np.sqrt(var / 5)
    
plt.hist(sample_5, bins=50, normed=True)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')

x = np.linspace(0,4,100)
pdf = sts.norm(mu_5, sigma_5).pdf(x)
plt.plot(x, pdf, lw=3, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
