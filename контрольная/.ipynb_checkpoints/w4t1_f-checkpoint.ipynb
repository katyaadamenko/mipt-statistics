# Центральная предельная теорема


## 1. Для оценки работы центальной предельной теоремы в качестве примера возьмем непрерывное распределение Хи-квадрат

```python
#для начала работы импоритирем необходимые модули
```

```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#КОД, ГЕНЕРИРУЮЩИЙ  ИСХОДНУЮ ВЫБОРКУ
#сгенерируем выборку объемом n=1000 для выбранного непрерывного распределения, а именно Хи-квадрат
chi2_rv = sts.chi2(1)
n = chi2_rv.rvs(1000)
```

```python
#КОД ДЛЯ ПОСТРОЕНИЯ ИСХОДНОГО ГРАФИКА
# для визуализации характера распределения построим график, иллюстирующий распределение случайной величины и плотность ее распределения
plt.hist(n, normed=True, label = 'Distribution rv')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')

x = np.linspace(0,10,1000)
pdf = chi2_rv.pdf(x)
plt.plot(x, pdf, label='PDF')
plt.legend(loc='upper right')
plt.show()
```

# 2. ЦПТ утверждает, что сумма n независимых одинаково распределённых случайных величин имеет распределение, близкое к нормальному. Рассмотрим распределение случайной величины для выборки n=1000, но для случая, если каждое случайное значение является средней из k наблюдений. k будет принимать значения 5, 10 и 50

```python
#КОД, ГЕНЕРИРУЮЩИЙ ВЫБОРКИ ДЛЯ ОЦЕНКИ РАСПРЕДЕЛЕНИЯ ВЫБОРОЧНОГО СРЕДНЕГО С ШАГОМ k
# сгенерируем выборки из средних
k5=5
k10=10
k50=50
n5 = [np.mean(chi2_rv.rvs(k5)) for i in range(1000)]
n10 = [np.mean(chi2_rv.rvs(k10)) for i in range(1000)]
n50 = [np.mean(chi2_rv.rvs(k50)) for i in range(1000)]
```

```python
#КОД, ГЕНЕРИРУЮЩИЙ ПОСТРОЕНИЕ ГИСТРОГРАММ ВЫБОРОК ДЛЯ ОЦЕНКИ РАСПРЕДЕЛЛЕНИЯ ВЫБОРОЧНОГО СРЕДНЕГО С ШАГОМ k
for i in(n5, n10, n50):
    plt.hist(n5, normed=True, label = 'Distribution for n from k')
    plt.ylabel('$F(x)$')
    plt.xlabel('$x$')
    plt.legend(loc='upper right')
    plt.show()
```

```python
#КОД, ГЕНЕРИРУЮЩИЙ ПАРАМЕТРЫ НОРМАЛЬНЫХ РАСПРЕДЕЛЕНИЙ ДЛЯ АППРОКСИМАЦИИ ПРИ РАЗЛИЧНЫХ k
m=np.mean(n) # средняя исхдного распределения
d = np.var(n)
SKO=np.std (n) # среднеквадратическое отклонение исходного распределения
# рассчитаем параметры аппроксимации, а именно СКО (s5, s10, s50 ), m5=m10=m50=m
s5 = SKO/(k5**0.5)
s10= SKO/(k10**0.5)
s50= SKO/(k50**0.5)
print ("Средняя=", m)
print ("Дисперсия исходной выборки =", d)
print ("Для шага 5: Дисперсия =", d/k5, "СКО=",s5)
print ("Для шага 10: Дисперсия =", d/k10, "СКО=",s10)
print ("Для шага 50: Дисперсия =", d/k50, "СКО=",s50)
```
